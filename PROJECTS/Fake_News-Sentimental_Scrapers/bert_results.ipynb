```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
epochs = range(1, 6)
```

```python
bert_regexp = {
    "accuracy": [0.98389, 0.98942, 0.99062, 0.99207, 0.99255],
    "f1_score": [0.93109, 0.93581, 0.93806, 0.94238, 0.94340],
    "val_loss": [0.08510, 0.06404, 0.06901, 0.06678, 0.06457],
}

bert_regexp_stopwords = {
    "accuracy": [0.96274, 0.98341, 0.98702, 0.98582, 0.98726],
    "f1_score": [0.89888, 0.93199, 0.94005, 0.93588, 0.93480],
    "val_loss": [0.22811, 0.08396, 0.08397, 0.09248, 0.11539],
}

bert_regexp_stopwords_lemmatization = {
    "accuracy": [0.98269, 0.99207, 0.99111, 0.99327, 0.99375],
    "f1_score": [0.92962, 0.94342, 0.94340, 0.94412, 0.94318],
    "val_loss": [0.09254, 0.04514, 0.05414, 0.05169, 0.04326],
}
```

```python
f = plt.figure()
f.set_figwidth(6)
f.set_figheight(4)

plt.plot(epochs, bert_regexp["accuracy"])
plt.plot(epochs, bert_regexp_stopwords["accuracy"])
plt.plot(epochs, bert_regexp_stopwords_lemmatization["accuracy"])
plt.title("Accuracy Per Epoch", fontsize=20)
plt.xlabel("Epoch", fontsize=16)
plt.xticks(fontsize=14, ticks=epochs)
plt.ylabel("Accuracy", fontsize=16)
plt.ylim([0.96, 1])
plt.yticks(fontsize=14)
plt.legend(["regexp", "regexp+stop", "regexp+stop+lemma"], loc="lower right")
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(6)
f.set_figheight(4)

plt.plot(epochs, bert_regexp["f1_score"])
plt.plot(epochs, bert_regexp_stopwords["f1_score"])
plt.plot(epochs, bert_regexp_stopwords_lemmatization["f1_score"])
plt.title("F1 Score Per Epoch", fontsize=20)
plt.xlabel("Epoch", fontsize=16)
plt.xticks(fontsize=14, ticks=epochs)
plt.ylabel("F1 Score", fontsize=16)
plt.ylim([0.99 * min(bert_regexp_stopwords["f1_score"]), 1.01 * max(bert_regexp["f1_score"])])
plt.yticks(fontsize=14)
plt.legend(["regexp", "regexp+stop", "regexp+stop+lemma"], loc="lower right")
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(6)
f.set_figheight(4)

metric = "val_loss"

plt.plot(epochs, bert_regexp[metric])
plt.plot(epochs, bert_regexp_stopwords[metric])
plt.plot(epochs, bert_regexp_stopwords_lemmatization[metric])
plt.title("Validation Loss Per Epoch", fontsize=20)
plt.xlabel("Epoch", fontsize=16)
plt.xticks(fontsize=14, ticks=epochs)
plt.ylabel("Val Loss", fontsize=16)
plt.ylim([0.9 * min(bert_regexp_stopwords_lemmatization[metric]), 1.1 * max(bert_regexp_stopwords[metric])])
plt.yticks(fontsize=14)
plt.legend(["regexp", "regexp+stop", "regexp+stop+lemma"], loc="upper right")
plt.show()
```
